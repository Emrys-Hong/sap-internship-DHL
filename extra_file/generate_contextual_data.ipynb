```python
import pandas as pd
import random
```

```python
raw_txt = pd.read_excel('parcel_data.xlsx')
```

```python
raw_txt.head()
```

```python
random_index = list(raw_txt.index)
random.shuffle(random_index)
```

```python
with open('training_dataset_direct_train.txt','w',encoding='utf-8') as f:    
    for index in random_index[:int(len(raw_txt.index)*0.6)]:
        mylist = []
        try:
            string = ''
            names = raw_txt.loc[index, 'Recipient name'].split()
            for name in names:
                string += name
                string += ' name\n'
        except Exception:
            pass
        mylist.append(string)
        try:
            string = ''
            addresses = raw_txt.loc[index, 'AddressLine1'].split()
            for address in addresses:
                string += address
                string += ' address\n'
        except Exception:
            pass
        mylist.append(string)
        try:
            string = ''
            districts = raw_txt.loc[index, 'อำเภอ District'].split()
            for district in districts:
                string += district
                string += ' district\n'
        except Exception:
            pass
        mylist.append(string)
        try:
            string = ''
            states = raw_txt.loc[index, 'Recipient state'].split()
            for state in states:
                string += state
                string += ' state\n'
        except Exception:
            pass
        mylist.append(string)
        try:
            string = ''
            zipcodes = str(raw_txt.loc[index, 'Recipient zipcode']).split()
            for zipcode in zipcodes:
                string += zipcode
                string += ' zipcode\n'
        except Exception as e:
            pass
        mylist.append(string)
        try:
            string = ''
            phones = str(raw_txt.loc[index, 'Recipient phone']).split()
            for phone in phones:
                string += phone
                string += ' phone\n'
        except Exception:
            pass
        
        ## randomly shuffle the string
        mylist.append(string)
        random.shuffle(mylist)
        
        for string in mylist:
            f.write(string)
        f.write('\n')
```
